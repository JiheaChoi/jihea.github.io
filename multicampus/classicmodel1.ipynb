```python
import pymysql.cursors
```

```python
import pandas as pd
```

```python
connection = pymysql.connect(host='localhost', 
        user='root',
        password='1111',
        db='classicmodels',
        charset='utf8',
        cursorclass=pymysql.cursors.DictCursor) # localhost

sql = 'select customers.customerNumber, customers.salesRepEmployeeNumber, payments.checkNumber from customers LEFT JOIN payments on customers.customerNumber = payments.customerNumber;'

try:
    with connection.cursor() as cursor:
        # Read a single record
        #sql = "SELECT `id`, `password` FROM `users` WHERE `email`=%s"
        #sql = "SELECT total_bill FROM tip.tips where tip >= 7 ;"
        #sql = "SELECT count(total_bill) FROM tip.tips where tip >= 7 ;"
        cursor.execute(sql)
        result = cursor.fetchall()
        df= pd.DataFrame(result)
        #print(result)
finally:
    connection.close()
```
