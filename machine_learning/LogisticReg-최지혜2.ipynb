```python
import pandas as pd

csv_path = "Pro2.csv"   #csv 파일 이름

def load_data(data_path=csv_path):  #파일 읽는 함수
    return pd.read_csv(data_path)

load_data()  #파일 읽기 
```

```python
Passenger = load_data()  
Passenger.head()  
```

```python
list(Passenger.keys()) #탑승객 데이터 list
```

```python
from sklearn import datasets           #Iris Virginica 예제
from sklearn.linear_model import LogisticRegression
import numpy as np

iris = datasets.load_iris()

log_reg = LogisticRegression()

X = iris["data"][:,3:]
y = (iris["target"]==2).astype(np.int)

log_reg.fit(X,y)
```

```python
import matplotlib.pyplot as plt
X_new = np.linspace(0,3,1000).reshape(-1,1)
y_proba = log_reg.predict_proba(X_new)


# 추정확률 plotting
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris virginica")
decision_boundary = X_new[y_proba[:,1]>=0.5][0]

plt.plot([decision_boundary, decision_boundary],[-1,2], "r:", linewidth=2)

```
