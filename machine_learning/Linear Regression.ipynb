```python
import pandas as pd

csv_path = "키_몸무게.csv"   #csv 파일 이름

def load_data(data_path=csv_path):  #파일 읽는 함수
    return pd.read_csv(data_path)

load_data()  #파일 읽기 
```

```python
height_weight = load_data()  
height_weight.head()   #head() 사용하기 (문제 1번)
```

```python
import matplotlib.pyplot as plt
height_weight.hist(bins=50, figsize=(20,15))
plt.show()   #데이터 시각화 하기 (문제2번)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X = height_weight[['height']]   # X를 height열로 정의
y = height_weight['weight']     # y를 weight열로 정의

lin_reg = LinearRegression()    #선형회귀 구현
lin_reg.fit(X,y)
lin_reg.intercept_, lin_reg.coef_   #절편과 기울기 출력 (문제 3번)
```

```python
lin_reg.predict([[160]])  #키가 160인 사람의 몸무게 예측 (문제 3번)
```

```python
plt.plot(X,lin_reg.predict(X),"r-")   #예측결과 시각화하기 (문제 4번)
plt.plot(X,y,"b.")
plt.axis([145,180,0,80])
plt.show()
```
